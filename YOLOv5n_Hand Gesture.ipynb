```python
model_path = '/content/runs/detect/train2/weights/best.pt'

# Load model dan ekstrak metrik validasi
model = YOLO(model_path)
val_results = model.val()
metric = val_results.box

precision = sum(metric.p) / len(metric.p)
recall = sum(metric.r) / len(metric.r)
F1_score = sum(metric.f1) / len(metric.f1)
mAP_05 = metric.map50
mAP_05_095 = metric.map

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {F1_score:.4f}")
print(f"mAP@0.5: {mAP_05:.4f}")
print(f"mAP@0.5:0.95: {mAP_05_095:.4f}")

# Grafik metrik tunggal (contoh)
import matplotlib.pyplot as plt
import numpy as np

labels = ['Precision', 'Recall', 'mAP@0.5', 'mAP@0.5:0.95', 'F1-score']
scores = [precision, recall, mAP_05, mAP_05_095, F1_score]

plt.figure(figsize=(8,5))
plt.bar(labels, scores, color='skyblue')
plt.ylim(0,1)
plt.title('Evaluation Metrics for Model')
plt.show()
```
